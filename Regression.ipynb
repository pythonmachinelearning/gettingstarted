```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Creae a dataframe from the data file. (i.e. csv)

```python
df = pd.read_csv('clean_df_automobiles.csv', index_col=0)
```

```python
df = df._get_numeric_data()
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df['horsepower'], df['price'])
```

```python
import seaborn as sns
```

```python
sns.regplot(x='horsepower', y='price', data=df)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
X = df[['horsepower']] # independent variable
```

```python
Y = df['price'] # dependent variable
```

```python
lm.fit(X, Y) # train the LinearRegression Model
```

```python
Yhat = lm.predict(X) # Predict the target/independent variable Y
```

```python
df['predictions'] = pd.Series(Yhat) # add the predicted values to the dataframe
```

```python
df[['horsepower', 'price', 'predictions']] # show ground truth and predicted labels together in dataframe
```

```python
width = 12
height = 10
plt.figure(figsize=(width, height))

ax1 = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
sns.distplot(df['predictions'], hist=False, color="b", label="Fitted Values" , ax=ax1)

plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Cars')

plt.show()
plt.close()
```
